**Step 2: Import the data and display the head, shape, and description of the data.**

```python
import pandas as pd

# Importing the data
data = pd.read_csv('data.xlsx - Sheet1.csv')

# Displaying the head of the data
print("Head of the data:")
print(data.head())

# Displaying the shape of the data
print("\nShape of the data:")
print(data.shape)

# Describing the data
print("\nDescription of the data:")
print(data.describe())
```

**Step 3: Univariate Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(data['Salary'], bins=20, kde=True)
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

```

**Step 4: Bivariate Analysis
Relationship between numerical columns**

```python
# Scatter plot between '10percentage' and 'Salary'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='10percentage', y='Salary', data=data)
plt.title('Scatter plot between 10th Percentage and Salary')
plt.xlabel('10th Percentage')
plt.ylabel('Salary')
plt.show()
# Observations: Add observations here after each plot.

# Similar analysis for other numerical columns.
```

**Patterns between categorical and numerical columns**

```python
# Boxplot between 'Gender' and 'Salary'
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Salary', data=data)
plt.title('Boxplot of Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()
# Swarmplot between 'Specialization' and 'Salary'
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Specialization', y='Salary', data=data)
plt.title('Swarmplot of Salary by Specialization')
plt.xlabel('Specialization')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()

```

**Relationships between categorical columns**

```python
# Stacked bar plot between 'Gender' and 'Specialization'
gender_specialization = pd.crosstab(data['Gender'], data['Specialization'])
gender_specialization.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Plot of Specialization by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

```
